```python
# Fungsi untuk menghapus titik lebih dari satu
def hapus_titik(df):
   df= re.sub(r'\.{2,}', '.', df)
   df=re.sub(r'\b\d{16}\b','', df) # menghapus angka 16 untuk KK
   df= re.sub(r'[+]?[0-9]+', '', df) # menghapus angka
   df=re.sub(r'\,<>./?@#$%^&*_~','',df)
   return df
```

```python
# Dictionary of abbreviations
abbreviation_dict = {
    "KK": "Kartu Keluarga",
    'KIP': "Kartu Indonesia Pintar",
    'KIS': "Kartu Indonesia Sehat",
    'BST': "Bantuan Sosial Tunai",
    'PKH': "Program Keluarga Harapan",
    'PBI': "Penerima Bantuan Iuran",
    'DINSOS': "Dinas Sosial",
    'BANSOS': "Bantuan Sosial",
    'KTP': "Kartu Tanda Penduduk",
    'BPNT': "Bantuan Pangan Non Tunai",
    'BLT' : "Bantuan Langsung Tunai",
    'NIK': "No Induk Kependudukan"
}
```

```python
def expand_abbreviations(text, abbreviation_dict):
    # Check if the input is a string; if not, return it unchanged
    if not isinstance(text, str):
        return text
    
    # Split the text into words
    words = text.split()
    
    # Replace abbreviations with their expanded forms using the provided dictionary
    expanded_words = []
    for word in words:
        # Check if the word exists in the abbreviation dictionary; if so, replace it
        expanded_word = abbreviation_dict.get(word, word)
        expanded_words.append(expanded_word)
    
    # Join the expanded words back into a single string
    return ' '.join(expanded_words)


# Terapkan fungsi ke kolom yang sesuai
df['aduan'] = df['aduan'].apply(lambda x: expand_abbreviations(x, abbreviation_dict))
df['pertanyaan'] = df['pertanyaan'].apply(lambda x: expand_abbreviations(x, abbreviation_dict))
df['jawaban'] = df['jawaban'].apply(lambda x: expand_abbreviations(x, abbreviation_dict))

# Cetak hasil untuk memeriksa
print(df[['aduan', 'pertanyaan', 'jawaban']])

```

```python
import random
replacement_word =["Bagaimana", "apakah","gimana","kenapa","mohon","segera","bisakah"]

def replace_random_word(pertanyaan, replacement_word):
    words=pertanyaan.split()
    
    if len(words)>0:
        random_word_idx= random.randint(0,len(words)-1)

        random_replacement=random.choice(replacement_word)
        words[random_word_idx]=random_replacement
    
    return"".join(words)

# 3. Terapkan fungsi untuk mengganti kata acak pada setiap teks di kolom 'text'
df['modified_text'] = df['pertanyaan'].apply(lambda x: replace_random_word(x, replacement_word))

# 4. Tampilkan hasilnya
print(df)
```
